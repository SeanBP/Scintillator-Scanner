```python
import matplotlib.pyplot as plt
import pylab
import numpy as np
import mplhep as hep
hep.style.use(hep.style.CMS)
hep.set_style("CMS")


#analysis with blips

filenumber = 961
avgp2p_withblips = []
    
for i in range(0, filenumber):                             #iterate through files
    
    peaktopeak_withblips = []
    
    with open("../data/Jun14Scan/"+str(i + 1) + ".txt") as f:
        current_file = (f.read().split('-- Event'))
        
    
    for j in range(0, len(current_file)):                  #iterate through events
        lines = current_file[j].split('\n')
        maxvoltage = 0;
        minvoltage = 0;
        baseline = []
        for line in lines[400:500]:
            values = line.split()
            baseline.append(float(values[3]))
        if len(baseline) > 0:
            base = sum(baseline) / len(baseline)
        else:
            base = 0
        for line in lines[500:700]:
            values = line.split()
            
            if float(values[3]) < minvoltage:              #get p2p values
                minvoltage = float(values[3])
            #if float(values[3]) > maxvoltage:
                #maxvoltage = float(values[3])
            maxvoltage = base
        peaktopeak_withblips.append(maxvoltage - minvoltage)
    avg_withblips = sum(peaktopeak_withblips) / len(peaktopeak_withblips)
    if avg_withblips > 0 and avg_withblips < 2000:
        avgp2p_withblips.append(avg_withblips/1.6)
    elif avg_withblips < 0:
        avgp2p_withblips.append(0)
    elif avg_withblips > 2000:
        avgp2p_withblips.append(2000)
    if i%50 == 0:
        print(str(round(100 * (i/filenumber)))+"%")
        

        
#For square heatmap


rows = int(filenumber / np.sqrt(filenumber))
columns = int(filenumber / np.sqrt(filenumber))
heatmap = [0]*rows

k = 0
for i in range(0, rows):                                   #make 2d array of averages
    column = []
    for j in range(0, columns):
        column.append(avgp2p_withblips[k])
        k += 1
    heatmap[i] = column
print(heatmap)
```

```python
from matplotlib import colors
plt.imshow(heatmap, cmap='viridis',vmin = 0, extent=[0,7.5,7.5,0])
#norm=colors.LogNorm()
plt.title("Tile Scan")
plt.xlabel("x position (cm)")
plt.ylabel("y position (cm)")
cbar = plt.colorbar()
cbar.set_label("Photoelectrons")
plt.show
```
